from above graphs, we conclude that most of the buyers are female as well as amount spent by females are also greater.

# state

```python
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.barplot(x='State', y='Orders', data= sales_state)
# sns.set(rc={'figure.figsize':(15,5)})
```

```python
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='State', y='Amount', data=sales_state)
```

From above graphs, we can conclude that most of the orders are from UP, Maharashtra and Karnataka as well as most amount spent are also by these states.

# Marital_Status

```python
t=sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(5,5)})
for bars in t.containers:
    t.bar_label(bars)
```

```python
sales_mar=df.groupby(['Marital_Status','Sex'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Marital_Status', y='Amount', data=sales_mar, hue='Sex')
```

we conclude that married women spent the most amount.

# Occupation

```python
t=sns.countplot(data=df, x='Occupation')
for bars in t.containers:
    t.bar_label(bars)
sns.set(rc={'figure.figsize':(20,5)})
```

```python
sales_ocp=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Occupation', y='Amount', data=sales_ocp)
```

most buyers are from  IT, healthcare and then aviation and aslo most sales are respectively.

# Product Category

```python
t=sns.countplot(data=df, x='Product_Category')
for bars in t.containers:
    t.bar_label(bars)
sns.set(rc={'figure.figsize':(25,5)})
```

```python
sales_prd=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x='Product_Category', y='Amount', data=sales_prd)
sns.set(rc={'figure.figsize':(25,5)})
```

Most number of sales are from category clothes, food and then electronics but most revenues are generated from food, clothing and then electronics.
